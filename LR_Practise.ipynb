```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.metrics  import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
br_data = sklearn.datasets.load_breast_cancer()
```

```python
br_data
# so data is there , target is there
```

```python
df = pd.DataFrame(br_data.data , columns = br_data.feature_names)
```

```python
df
```

```python
df.columns
```

```python
# insert target column
df["label"] = br_data.target
```

```python
df.columns
```

```python
df['label'].value_counts().sum()
```

```python
df.groupby('label').mean()
```

```python
df.isnull().sum()
```

```python
df['label'].value_counts().count()
```

```python
df['label'].value_counts()
```

```python
# label diff is somewhat ok
# we will split
X = df.drop(columns='label' , axis = 1)
Y= df['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
# 0 label is malignant
# 1 is benign
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , random_state = 1 , test_size = 0.2)
```

```python
X.shape , X_train.shape , X_test.shape
```

```python
# model training
model = LogisticRegression()
```

```python
model.fit(X_train , Y_train)
```

```python
tr_predict = model.predict(X_train )
```

```python
tr_accuracy = accuracy_score(Y_train, tr_predict)
```

```python
print("Acuurcy of trainingg data :   " , tr_accuracy)
```

```python
ts_predict = model.predict(X_test)
```

```python
ts_accuracy = accuracy_score(Y_test , ts_predict)
```

```python
print("Accuracy of Testing data :  ", ts_accuracy)
```
