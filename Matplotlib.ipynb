```python
#importing matplotlib
import matplotlib.pyplot as plt
```

```python
# import numpy  to get data for our plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(y)
```

```python
print(z)
```

```python
# Plotting the data
# sine wave because y is the sin value of x
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
# cosine wave , cause z is x's cos value
```

```python
# adding title for axeses
plt.plot(x,y)
plt.xlabel('random Numbers')
plt.ylabel('sin value')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x , np.cos(x), 'r--')
plt.show()
```

```python
# Bar graph

fig = plt.figure() # creates empty plot
ax = fig.add_axes([0,0,1,1])  # 2 0's r origin , 2 1's r height & weight of rectangle
languages = ['English','Spanish','Latin','French','German']
people = [20,30,50,10,60]
ax.bar(languages, people) # bar  graph plotting and declaring what var to show
plt.xlabel('Languages')
plt.ylabel('No of People')
plt.show()
```

```python
# Pie chart
fig1 = plt.figure() # creates empty plot
ax = fig1.add_axes([0,0,1,1])  # 2 0's r origin , 2 1's r height & weight of rectangle
languages = ['English','Spanish','Latin','French','German']
people = [20,30,50,10,60]
ax.pie( people, labels = languages , autopct= '%1.1f%%') # bar  graph plotting and declaring what var to show
# in pie chart , only 1 axis , not x, y
# so we will use only what we want as op , that is people
# labels will be languages , autopct is , how many floating pt value we want
# that is %1.1f%%' , gives one value after decimal
plt.show()
```

```python
# Scatter plot
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color= 'g')
ax.scatter(x,z,color= 'b')
plt.show()
```

```python
# 3D Scatter plot
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
