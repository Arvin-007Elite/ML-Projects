```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# loading dataset
dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(dataset)
```

```python
# this tells , whether the person is in benign or  worst state
```

```python
# loading the data to pandas df
df =  pd.DataFrame(dataset.data, columns=dataset.feature_names)
# data is var which holds all data in array in dataset
# columns is var in datset which holds  all columns
# we didnt include target variables as op
```

```python
df.head()
```

```python
# some datas r bw 10 to 50 , in decimals , in 1000s , in 100s
# here only will perform data standardization  , for bettr performance
```

```python
df.shape
```

```python
# now will store all input values in X
# and store all target vars (op) in Y
```

```python
X = df
Y = dataset.target
```

```python
print(X)
```

```python
print(Y)
```

```python
# Splitting Trainig data , and Testing data

X_train,X_test,Y_train,Y_test = train_test_split(X,Y , test_size=0.2, random_state=3)
# 0.2 is 20% of data for testing
# random state is 3 means reproducing the loops 3 times

# if a data column has values in 1000s , but 1 value alone in 100 , it is called an outliers , so we use standardization

```

```python
# lets check the splitting amount

print(X.shape , X_train.shape , X_test.shape)
```

```python
# Now Standardize the data
print(dataset.data.std())

# data std must be 1 , if  all datas r at same range
```

```python
# the above op tells that , dtas r not in same range , and varies a lot

# for this will use Standard scaler
```

```python
# Standardize features by removing the mean and scaling to unit variance.

# The standard score of a sample x is calculated as:

# z = (x - u) / s

# where u is the mean of the training samples or zero if with_mean=False, and s is the standard deviation of the training samples or one if with_std=False.

# Centering and scaling happen independently on each feature by computing the relevant statistics on the samples in the training set. Mean and standard deviation are then stored to be used on later data using transform.

# Standardization of a dataset is a common requirement for many machine learning estimators: they might behave badly if the individual features do not more or less look like standard normally distributed data (e.g. Gaussian with 0 mean and unit variance).
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
# now we have  to transform our data
X_train_standardized = scaler.transform(X_train)
```

```python
print(X_train_standardized)
```

```python
# now every values r -1 to 1 ,
# but now we have to standardized X_Test also
# but we have very less dataset  in X_test , we shouldnt do fit() for teat
# we should just transform based on train
```

```python
X_test_standardized = scaler.transform(X_test)
```

```python
print(X_train_standardized.std())
```

```python
# now its in crct range , but it doesnt affect the dats nature

print(X_test_standardized.std())

```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Label encoding of breast cancer datset
cancer_data = pd.read_csv('/content/wisc_bc_data.csv')
```

```python
cancer_data.head()
# M - Malignant (end stage)
# B - Benign (early stage)
```

```python
# Finding count of Diff labels
cancer_data['diagnosis'].value_counts()
```

```python
# Load Label encoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis)
# this will create new col called labels
```

```python
# appending labels to the dataframe
cancer_data['target'] = labels
```

```python
cancer_data.head()
```

```python
# new col added
cancer_data['target'].value_counts()
```
