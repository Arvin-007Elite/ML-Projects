```python
# Seaborn - Data Visualization Library
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# loading seaborn inbuilt datset
# tips is a datset of bill
tips = sns.load_dataset('tips')
```

```python

tips.head()
```

```python
# visualize
# relplot gives u 2 plots
sns.relplot(data=tips, x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size')
plt.show()
```

```python
# we have made the total bill as x , tip as y , time has 2 types
# dinner and lunch , so grapdh will be shown for 2 cases separtly
# hue and style is for smoker , cause if smoker , it shows isin diff color and shape
# size is set as size col in datset cause , size is no of persons!
# if 1 member means plotted dot shape will  differ accord with size increses
# Based on which col u give in hue , that only differentiates in the graph
```

```python
# in matplotlib we have to mention hue and style color
# but in seaborn , it is not needed !

# we can change the theme for plots
sns.set_theme()
sns.relplot(data=tips, x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size')
plt.show()

# above plot and below plot design have changed
```

```python
# Loading Titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
# datset tells persons survived from titanic crash , 0 means not survived , 1 means survived
```

```python
# count plot
# to find the no of pts in each parameters
```

```python
sns.countplot(x='class',data=titanic)
```

```python
titanic.shape
```

```python
#  finding how many survived
sns.countplot(x='survived',data=titanic)
```

```python
# Bar chart
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```

```python
# Distribution plot
sns.displot(titanic['class'])
# titanic.head()
# displays  the distribution of seats by classes
```
