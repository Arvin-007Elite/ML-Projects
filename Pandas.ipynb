```python
# Correlation:
# 1. Positive ( if onlyvalue increse , other also increases ) (positivly corelated)
# 2. Negative means ( in a city if crime rate increases , tax will decrease ) (negatively corelated)
# example it as a col name in  dataset

h_price = boston_df.corr()

# each col is compared with other col
# if ur wanted col is negative when compred to other , it is negative corr
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(h_price, cbar=True, square = True , fmt = '.1f', annot=True , annot_kws={'size':8},cmap='Blues')
# cbar is the color bar in heat map ,
#  fmt  is no of floating pt , 1 value after decimal
# cmpa is color of map is blue , annot is annotation and its size
```

```python
# if the color is light , then expected value or correlation is negative
# if dark , it is positivly correlated
```
