```python
# both files has country code , we can combine it and see
final_df = pd.merge(df,df_c, on='Country Code', how='left')
final_df
```

```python
Country_names = final_df.Country.value_counts().index
Country_names
# will get only the indexes
```

```python
 Country_val = final_df.Country.value_counts().values

 # will get only the values of the value count's indexes
```

```python
# to see which country has more usage
plt.pie(Country_val, labels=Country_names)
```

```python
# to see which country has more usage for only Top 3 countries
plt.pie(Country_val[:3], labels=Country_names[:3], autopct="%1.1f%%")
```

```python
# Observation:
# Zomato's maximum records or Tramsactions are from India , Then USA & UK
```

```python
# now will see which country gives more rating
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Counts'})
```

```python
ratings
```

```python
# here 2148 members didint give ratings ! , that is maximum !

# whenver 4.5 to 4.9 rating it indicates that food delivery was exceelnt
# fom 4.0 to 4.4 ratings , it is very good
# if bw 3.5 to 3.9 , it is good
# if bw 3.0 to 3.4 , its average
# when the ratings is bw 2.5 to 2.9 ok average
#  it is 2.0 to 2.4 , its below averageratings
```

```python
ratings.head()
```

```python
import matplotlib

# Set figure size
# plt.figure(figsize=(12, 6))
# (or)
matplotlib.rcParams['figure.figsize'] = (12,6)
# Create the bar plot
sns.barplot(x="Aggregate rating", y="Rating Counts",hue="Rating color", data=ratings,palette=['white','red','orange','yellow','green','grey'])
```

```python
# Observation:
# 1. Not rated count is very high
# 2. Maximum num of ratikgs r bw 2.5 to 3.4
```

```python
# Count plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','grey'])
```
