```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/2021_Tamil_Nadu_Legislative_Assembly_election#:~:text=The%20Dravida%20Munnetra%20Kazhagam%20(DMK,He%20replaced%20Edappadi%20K'
```

```python
src = requests.get(url)
```

```python
soup = BeautifulSoup(src.text , 'html')
```

```python
print(soup)
```

```python
print(soup.prettify())
```

```python
soup.find('div')
```

```python
soup.find_all('div')[1]
```

```python
soup.find_all('div' , class_='vector-header-start')
```

```python
table1 = soup.find_all('table' , class_= 'wikitable')[8]
```

```python
total_voters = table1.find_all('th')
```

```python
total_voters
```

```python
total_voters_21 = [vote.text.strip() for vote in total_voters]
total_voters_21
```

```python
import pandas as pd
df  = pd.DataFrame(columns=total_voters_21)
```

```python
df
```

```python
column_data = table1.find_all('tr')
```

```python
column_data
```

```python
for row in column_data[1:]:
  row_data = row.find_all('td')
  ind_row_data = [row.text.strip() for row in row_data]
  # print(ind_row_data)
  if len(ind_row_data) == len(total_voters_21):
    df.loc[len(df)] = ind_row_data
```

```python
# df.drop(0 , inplace=True)
print(ind_row_data)
```
